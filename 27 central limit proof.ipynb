```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/batter.csv')
runs = df['runs'].values
sns.kdeplot(runs,)
```

```python
m = 500
n = 60
new = []
for i in range(m):
   new_arr= np.random.choice(runs, size=n)
   new.append(new_arr.mean())
   
   
new = np.array(new) 
```

```python
sns.kdeplot(new,color='red')
```
