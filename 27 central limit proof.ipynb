```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/batter.csv')
runs = df['runs'].values
runs
```

```python
m = 500
n = 60
new = []
for i in range(m):
   new_arr= np.random.choice(runs, size=n)
   new.append(new_arr.mean())
   
   
new = np.array(new) 
```

```python
sns.kdeplot(new,color='red')
```

```python

print(runs.mean())
print(new.mean())
```

```python
df = sns.load_dataset('titanic')
df['age'].fillna(df["age"].mean(), inplace=True)
age_vals = df['age'].values
```

```python
no_of_samples = 50
no_of_trials = 1000
new =[]
for i in range(no_of_trials):
    new_arr = np.random.choice(age_vals, size=no_of_samples)
    new.append(new_arr.mean())

new = np.array(new)
print(new.mean())
print(age_vals.mean())
```

```python
fig, ax = plt.subplots(1,2 ,figsize=(15,5))
sns.kdeplot(new,ax=ax[0], color='blue')
sns.kdeplot(new,ax=ax[1], color='green')
```
