```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("datasets/titanic.csv")
df['Fare'].head()

```

```python
sns.kdeplot(df['Fare'])
```

```python
n = 30
stds = []
samples = []
for i in range(10):
   x = np.random.choice(df['Fare'].values,size=n)
   stds.append(x.std())
   samples.append(x.tolist())
   


    
```

```python
samples = np.array(samples)
```

```python
# find the value of x- bar
sample_mean = samples.mean()
sample_mean
```

```python
# find the standard deviation of the stds
sample_std = np.mean(stds)
sample_std
```

```python
lower_limit = sample_mean - 2.042*(sample_std/np.sqrt(30))
upper_limit = sample_mean + 2.042*(sample_std/np.sqrt(30))
```

```python
print(f"the range is between {lower_limit} - {upper_limit}")
```
